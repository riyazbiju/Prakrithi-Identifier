```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

# Load the dataset from a CSV file
data = pd.read_csv('/content/prakriti fnal number data.csv')
# Split the dataset into features and labels
X = data.drop('prakriti', axis=1)  # Features
y = data['prakriti']  # Labels

# Create the ANN model
model = MLPClassifier(hidden_layer_sizes=(100, 100), activation='relu', solver='adam')

# Train the model
model.fit(X, y)

# Function to predict prakriti based on user input
def predict_prakriti():
    options = {
        "bodytype": {1: "lean and weak", 2: "medium and fleshy", 3: "large and soft"},
        "face_shape": {1: "long, thin and small", 2: "moderate and sharp", 3: "large, round and fat"},
        "skintype_color": {1: "dry, rough and blackish", 2: "normal, soft and brown", 3: "oily, glossy and fair white"},
        "lips": {1: "dry, cracking and blackish", 2: "soft, thin and reddish", 3: "smooth, glossy and pinkish"},
        "palm": {1: "cracked, dry and black", 2: "moderate fat and brown", 3: "big, oily and firm"},
        "nail": {1: "thin, dry and break easily", 2: "moderate thick and yellow", 3: "thick and strong"},
        "tongue": {1: "thin or no coating and pink color", 2: "moderate coating and light pink color", 3: "thick coating and white yellowish color"},
        "feet": {1: "rough and cracked", 2: "reddish and soft", 3: "nothing specific"},
        "eye": {1: "dry, sunken, and dark circles", 2: "bright, clear and sharp", 3: "large, soft and good sight"},
        "sleep": {1: "interrupted (<6hrs)", 2: "sound (6-8hrs)", 3: "very sound (>6hrs)"},
        "appetite": {1: "variable, often and irregular", 2: "strong and regular", 3: "good and regular"},
        "voice": {1: "high-pitched, soft, and weak", 2: "loud, clear and sharp", 3: "deep and low pitch"},
        "emotion": {1: "easily stressed, anxious and fearful", 2: "easily angry, irritable and jealous", 3: "easily attached, sentimental"},
        "intolerance": {1: "cold", 2: "hot", 3: "can tolerate cold and heat"},
        "problem_solving": {1: "worrying constantly", 2: "can take bold decisions", 3: "can take right, firm decisions"},
        "hobbies": {1: "music, humor, and games", 2: "adventurous sports", 3: "nothing specific"},
        "memory": {1: "forgets quickly", 2: "moderate", 3: "good"},
        "grasping_power": {1: "sometimes quickly, sometimes weak", 2: "always quick and clever", 3: "always grasp late but understand best"},
        "fear": {1: "quick", 2: "rarely", 3: "good"},
        "anger": {1: "very quick", 2: "quick", 3: "less"},
        "friendship": {1: "few unstable", 2: "dependable", 3: "stable"},
        "sexual_desire": {1: "less", 2: "more", 3: "good energetic even after work"},
        "sweating": {1: "less sweating", 2: "more sweating", 3: "sweat very easily foul smell"},
        "urine": {1: "expels frequently large quantity and yellowish", 2: "moderate", 3: "good"},
        "immunity": {1: "less", 2: "good", 3: "good energetic even after work"},
        "strength": {1: "less tires quickly", 2: "can withstand physical or mental exertion", 3: "good"},
        "speech_and_voice": {1: "irrelevant speech", 2: "clear and loud", 3: "gentle, speak less"},
        "muscles": {1: "small, not well-built", 2: "medium", 3: "well-built"},
        "chest": {1: "unstable", 2: "loose/flaccid", 3: "steady"},
        "head": {1: "small", 2: "nothing specific", 3: "wide"},
        "forehead": {1: "coarse, thin, cracking", 2: "with folds", 3: "firm and strong"},
        "teeth": {1: "coarse thin cracking", 2: "medium with gap", 3: "firm and strong"}

    }

    user_input = {}  # Dictionary to store user input

    for feature, feature_options in options.items():
        print("your", feature + ":")
        for i, option in feature_options.items():
            print(i, ")", option)
        user_choice = int(input())
        user_input[feature] = user_choice

    # Count the occurrences of each selected option
    counts = {1: 0, 2: 0, 3: 0}
    for feature, value in user_input.items():
        counts[value] += 1

    # Find the majority selected option(s)
    majority_options = [option for option, count in counts.items() if count == max(counts.values())]

    # Calculate the percentage of each prakriti
    total_answers = sum(counts.values())
    percentages = {prakriti: count / total_answers * 100 for prakriti, count in counts.items()}

    # Print the prakriti and its percentage based on the majority selected option(s)
    if len(majority_options) == 1:
        if majority_options[0] == 1:
            print("The predicted prakrithi is: vata")
            print("Percentage Vata:", round(percentages[1], 2), "%")
            print("Percentage Pitta:", round(percentages[2], 2), "%")
            print("Percentage Kapha:", round(percentages[3], 2), "%")
        elif majority_options[0] == 2:
            print("The predicted prakrithi is: pitta")
            print("Percentage Vata:", round(percentages[1], 2), "%")
            print("Percentage Pitta:", round(percentages[2], 2), "%")
            print("Percentage Kapha:", round(percentages[3], 2), "%")
        elif majority_options[0] == 3:
            print("The predicted prakrithi is: kapha")
            print("Percentage Vata:", round(percentages[1], 2), "%")
            print("Percentage Pitta:", round(percentages[2], 2), "%")
            print("Percentage Kapha:", round(percentages[3], 2), "%")

    elif len(majority_options) == 2:
        if 1 in majority_options and 2 in majority_options:
            print("The predicted prakrithi is: vata and pitta")
            print("Percentage Vata:", round(percentages[1], 2), "%")
            print("Percentage Pitta:", round(percentages[2], 2), "%")
            print("Percentage Kapha:", round(percentages[3], 2), "%")
        elif 2 in majority_options and 3 in majority_options:
            print("The predicted prakrithi is: pitta and kapha")
            print("Percentage Pitta:", round(percentages[2], 2), "%")
            print("Percentage Kapha:", round(percentages[3], 2), "%")
            print("Percentage Vata:", round(percentages[3], 2), "%")
        elif 1 in majority_options and 3 in majority_options:
            print("The predicted prakrithi is: vata and kapha")
            print("Percentage Vata:", round(percentages[1], 2), "%")
            print("Percentage Kapha:", round(percentages[3], 2), "%")

            print("Percentage pitta:", round(percentages[3], 2), "%")
    else:
        print("The predicted prakrithi is: vata, pitta, and kapha")
        print("Percentage Vata:", round(percentages[1], 2), "%")
        print("Percentage Pitta:", round(percentages[2], 2), "%")
        print("Percentage Kapha:", round(percentages[3], 2), "%")

    # Create a pie chart of the predicted prakrithi percentages
    labels = ['Vata', 'Pitta', 'Kapha']
    sizes = [percentages[1], percentages[2], percentages[3]]
    colors = ['#ff9999', '#66b3ff', '#99ff99']
    explode = (0.1, 0, 0)  # explode the 1st slice

    plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('Predicted Prakrithi Percentages')
    plt.show()

# Call the function to predict prakriti based on user input
predict_prakriti()
```
